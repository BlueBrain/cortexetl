```python
import numpy
import pandas

from matplotlib import pyplot as plt
import seaborn as sns

sns.set(style="ticks", context="paper", font="Helvetica Neue",
        rc={"axes.labelsize": 7, "legend.fontsize": 6, "axes.linewidth": 0.6, "xtick.labelsize": 6, "ytick.labelsize": 6,
            "xtick.major.size": 2, "xtick.major.width": 0.5, "xtick.minor.size": 1.5, "xtick.minor.width": 0.3,
            "ytick.major.size": 2, "ytick.major.width": 0.5, "ytick.minor.size": 1.5, "ytick.minor.width": 0.3,
            "axes.titlesize": 7, "axes.spines.right": False, "axes.spines.top": False})

# SOURCES:
fn_S = "/gpfs/bbp.cscs.ch/project/proj83/home/ecker/sscx_connectivity_78hexes.npz"
# fn_C = "/gpfs/bbp.cscs.ch/project/proj83/scratch/bbp_workflow/sscx_calibration_mgfix/5-FullCircuit/5-FullCircuit-2-BetterMinis-Fpr15-StimScan-10x/bb16bd9f-3d21-4a35-8296-d6aec4c55bf7/4/corrs_78hexes.npz"
fn_Cs = ["/gpfs/bbp.cscs.ch/project/proj83/scratch/bbp_workflow/sscx_calibration_mgfix/5-FullCircuit/5-FullCircuit-2-BetterMinis-FprScan/5d83d4c2-693c-4ecc-a9da-c8dd2c8100c3/4/corrs_78hexes.npz", 
         "/gpfs/bbp.cscs.ch/project/proj83/scratch/bbp_workflow/sscx_calibration_mgfix/5-FullCircuit/5-FullCircuit-2-BetterMinis-Fpr15-StimScan-10x/bb16bd9f-3d21-4a35-8296-d6aec4c55bf7/4/corrs_78hexes.npz"]

for campaign_label, fn_C in zip(["spont", "evoked"], fn_Cs):

    # LOAD DATA:
    S = numpy.load(fn_S)["midrange"]
    C = numpy.load(fn_C)["corrs"]
    E = numpy.load(fn_C)["ei_corrs"]

    b = numpy.arange(0, 1.02, 0.02)
    Cbins = numpy.digitize(C, bins=b) - 1

    # ANALYSIS:
    distr = []
    for i in range(len(S)):
        Sc = numpy.sqrt(numpy.dot(S[:, i:i+1], S[:, i:i+1].transpose()))
        df = pandas.DataFrame({"struc": Sc.flatten(), "func": b[Cbins.flatten()]}).groupby("func").sum()
        distr.append(df["struc"] / df["struc"].sum())
    distr = pandas.concat(distr, axis=1, keys=range(len(S)))

    #PLOT:
    fig = plt.figure(figsize=(5, 3.5))

    ax1 = fig.add_axes([0.1, 0.25, 0.8, 0.6])
    ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.15])

    ax1.imshow(distr[numpy.argsort(E)], aspect="auto", extent=(0-0.5, len(E) - 0.5, 1.0, 0.0),
              clim=[0, 0.1], interpolation="nearest", cmap='viridis')
    ax1.set_ylabel("Corr. in long-range input")
    ax1.set_xticks([])
    ax1.invert_yaxis()

    ax2.plot(sorted(E))
    ax2.set_xlim(ax1.get_xlim())
    ax2.set_xticks([])
    ax2.set_xlabel("Hexes")
    ax2.set_ylabel("Internal corr.")

    fig.savefig("figures/nbS1_internal_vs_longrange_correlation_heatmap_" + campaign_label + ".pdf", bbox_inches='tight')

    fig = plt.figure(figsize=(2,2))
    ax = fig.gca()

    ax.plot(distr.apply(lambda x: (x.values * x.index.values).sum(), axis=0), E, 'o', c='k', ms=1)

    # p0 = [max(ydata), np.median(xdata),1,min(ydata)] # this is an mandatory initial guess
    # popt, pcov = curve_fit(sigmoid, xdata, ydata,p0, method='lm') ## Updated method from 'dogbox' to 'lm' 9.30.2021

    # ax.set_frame_on(False)
    ax.set_ylim([0.0, 1.0])
    ax.set_xlim([0.15, 0.4])
    ax.set_xlabel("Mean input correlation")
    ax.set_ylabel("Internal correlation")

    fig.savefig("figures/nbS1_internal_vs_longrange_correlation_scatter_" + campaign_label + ".pdf", bbox_inches='tight')
```
