```python
nrn_file = '/gpfs/bbp.cscs.ch/project/proj83/jira-tickets/NSETM-1948-extract-hex-O1/data/O1_data_physiology/S1nonbarrel_neurons/nodes.h5'
syn_file = '/gpfs/bbp.cscs.ch/project/proj83/jira-tickets/NSETM-1948-extract-hex-O1/data/O1_data/VPM__S1nonbarrel_neurons__chemical/edges.h5'

import h5py    
import numpy as np
nrns = h5py.File(nrn_file,'r+')
syns = h5py.File(syn_file,'r+')

layer_vals = np.array(nrns['nodes']['S1nonbarrel_neurons']['0']['layer'])
synapse_classes = np.array(nrns['nodes']['S1nonbarrel_neurons']['0']['synapse_class'])
syns_target_node_ids = np.array(syns['edges']['VPM__S1nonbarrel_neurons__chemical']['target_node_id'])
conductances = np.array(syns['edges']['VPM__S1nonbarrel_neurons__chemical']['0']['conductance'])

# print(np.min(syns_target_node_ids))

import matplotlib.pyplot as plt
fig, axes = plt.subplots(6, 2, figsize=(4, 12))
for synapse_class in [0, 1]:
    for layer in [0, 1, 2, 3, 4, 5]:
        
        ax = axes[layer][synapse_class]
        layer_5_nrn_ids = np.argwhere(np.logical_and(layer_vals == layer, synapse_classes==synapse_class)).flatten()

        all_conductances = []

        for nrn_id in layer_5_nrn_ids[:20]:
            synapses_onto_nrn = np.argwhere(syns_target_node_ids == nrn_id).flatten()
            for syn_id in synapses_onto_nrn:
                all_conductances.append(conductances[syn_id])

        ax.hist(all_conductances)
        ax.axvline(x = np.mean(all_conductances), color = 'b', label = str(np.mean(np.mean(all_conductances))))
        ax.legend()
        ax.get_yaxis().set_visible(False)
        
plt.savefig('conductances.pdf')
plt.show()
plt.close()
```
