```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def remove_spikes_from_membrane_potentials():

    a_soma.repo.report.df['value_spikes_removed'] = a_soma.repo.report.df['value']

    soma_window = a_soma.repo.report.df.etl.q(window="w1")
    spikes_window = a_spikes.repo.report.df.etl.q(window="w1")

    for simulation_id in a_soma.repo.simulations.df.simulation_id.unique():

        print(simulation_id)

        soma_sim_window = soma_window.etl.q(simulation_id=simulation_id)
        soma_spikes_window = spikes_window.etl.q(simulation_id=simulation_id)

        for i, gid in enumerate(soma_sim_window.gid.unique()):

    #         if gid not in regular_isi_gids:

            gid_soma = soma_sim_window.etl.q(gid=gid)
            gid_spikes = soma_spikes_window.etl.q(gid=gid)

            if len(gid_spikes.time):

                for st in gid_spikes.time:

                    inds_where_spike = gid_soma.index[(gid_soma.time > st - 3) & (gid_soma.time < st + 50)]
                    a_soma.repo.report.df.loc[inds_where_spike, 'value_spikes_removed'] = np.nan

    #             plt.figure()
    #             plt.plot(gid_soma.time, a_soma.repo.report.df.loc[gid_soma.index, "value_spikes_removed"])
    #             plt.show()
    #             plt.close()

    #         if i == 2000:
    #             break
    
def persist_df_with_spikes_removed_from_membrane_potentials():
    
    a_soma.repo.report.df.to_parquet('soma_report_with_spikes_removed.parquet')
    
    
def calculate_soma_features(df):

    groupby = ["neuron_class", "window", "simulation_id"]
    by_neuron_class = df.groupby(groupby, observed=True)["value_spikes_removed"].agg(["mean", "std"]).reset_index()
    return by_neuron_class


from scipy.stats import linregress
from scipy.ndimage import gaussian_filter1d

def calculate_pairwise_membrane_potential_correlations():

    mp_correlation_data = {'simulation_id': [], 'neuron_class': [], 'gid_0': [], 'gid_1': [], 'mp_corr_r': []}

    for simulation_id in a_soma.repo.simulations.df.simulation_id.unique():
        print(simulation_id)
        soma_sim_window = a_soma.repo.report.df.etl.q(simulation_id=simulation_id, window="w1")

        for neuron_class in a_soma.repo.report.df.neuron_class.unique():
            print(neuron_class)

            nc_soma = soma_sim_window.etl.q(neuron_class=neuron_class)
            uniq_gids = nc_soma.gid.unique()
            num_uniq_gids = len(uniq_gids)

            mp_corr_r = np.ones((num_uniq_gids, num_uniq_gids)) * np.nan
            mp_smoothed_corr_r = np.ones((num_uniq_gids, num_uniq_gids)) * np.nan

            for i in range(num_uniq_gids):
                for j in range(num_uniq_gids):
                    if i > j:
                        soma_i = nc_soma.etl.q(gid=uniq_gids[i])
                        soma_j = nc_soma.etl.q(gid=uniq_gids[j])

                        mp_correlation_data['simulation_id'].append(simulation_id)
                        mp_correlation_data['neuron_class'].append(neuron_class)
                        mp_correlation_data['gid_0'].append(uniq_gids[i])
                        mp_correlation_data['gid_1'].append(uniq_gids[j])

                        corr_r = linregress(soma_i.value, soma_j.value).rvalue
                        mp_corr_r[i, j] = corr_r
                        mp_correlation_data['mp_corr_r'].append(corr_r)

                        smoothed_analysis = False
                        if smoothed_analysis:
                            soma_i_value_smoothed = gaussian_filter1d(soma_i.value, 10)
                            soma_j_value_smoothed = gaussian_filter1d(soma_j.value, 10)
                            corr_smoothed_r = linregress(soma_i_value_smoothed, soma_j_value_smoothed).rvalue
                            mp_smoothed_corr_r[i, j] = corr_smoothed_r
                            mp_correlation_data['mp_smoothed_corr_r'].append(corr_smoothed_r)

        #             if (corr_r > 0.25):
        #                 plt.figure()
        #                 plt.plot(soma_i.time, soma_i.value)
        #                 plt.plot(soma_j.time, soma_j.value)
        #                 plt.show()
        #                 plt.close()

        #             break
        
#     break


    mp_correlation_df = pd.DataFrame.from_dict(mp_correlation_data)
    return mp_correlation_df

def persist_mp_correlation_df():
    mp_correlation_df.to_parquet('mp_correlation_df.parquet')
    
def calculate_mp_correlation_stats():
    groupby = ["neuron_class", "simulation_id"]
    mp_correlation_stats_by_nc_and_sim = mp_correlation_df.groupby(groupby, observed=True)["mp_corr_r"].agg(["mean", "std"]).reset_index()
```

```python
remove_spikes_from_membrane_potentials()
persist_df_with_spikes_removed_from_membrane_potentials()
soma_features_by_nc = calculate_soma_features(a_soma.repo.report.df)
calculate_pairwise_membrane_potential_correlations()
persist_mp_correlation_df()
calculate_mp_correlation_stats()
```
